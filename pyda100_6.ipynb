테크닉 51. 물류 데이터를 불러오자. 

```python
import pandas as pd 

factories = pd.read_csv("tbl_factory.csv", index_col=0) # 생산 공장 데이터 
factories
```

```python
warehouses = pd.read_csv("tbl_warehouse.csv", index_col=0) # 창고 데이터 
warehouses
```

```python
cost = pd.read_csv("rel_cost.csv", index_col=0)# 창고와 공장 간의 운송 비용
cost
```

```python
trans = pd.read_csv("tbl_transaction.csv", index_col=0) # 운송실적 
trans
```

```python
# 결합하기 
# 데이터 분석의 기초가 될 데이터는 운송 실적이므로 이를 중심으로 결합하자. 

join_data = pd.merge(trans, cost, left_on=["ToFC", "FromWH"], right_on=["FCID", "WHID"], how="left")
join_data
```

```python
# 공장 데이터 추가 
join_data = pd.merge(join_data, factories, left_on="ToFC", right_on="FCID", how="left")
join_data

# 공장ID(ToFC, FCID)를 조건으로 데이터 결합. 뒤에 공장 데이터 추가 
```

```python
# 창고 정보도 추가해서 직관적으로 보기 좋게 칼럼의 순서 재배치 
join_data = pd.merge(join_data, warehouses, left_on="FromWH", right_on="WHID", how="left")
join_data = join_data[["TransactionDate", "Quantity", "Cost", "ToFC", "FCName", "FCDemand", "FromWH", "WHName", "WHSupply", "WHRegion"]]
join_data
```
